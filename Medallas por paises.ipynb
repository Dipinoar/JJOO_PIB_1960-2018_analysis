```python
import pandas as pd

# Importar el archivo CSV
df = pd.read_csv('olympics_games.csv')

# Crear un diccionario para almacenar los totales de medallas
medals_totals = {}

# Iterar sobre cada fila del dataframe original y actualizar los totales de medallas
for index, row in df.iterrows():
    noc = row['NOC']
    sport = row['Sport']
    medal = row['Medal']
    
    if noc not in medals_totals:
        medals_totals[noc] = {}
    
    if sport not in medals_totals[noc]:
        medals_totals[noc][sport] = {
            'Gold': 0,
            'Silver': 0,
            'Bronze': 0,
            'Total': 0
        }
    
    if medal == 'Gold':
        medals_totals[noc][sport]['Gold'] += 1
    elif medal == 'Silver':
        medals_totals[noc][sport]['Silver'] += 1
    elif medal == 'Bronze':
        medals_totals[noc][sport]['Bronze'] += 1
    
    medals_totals[noc][sport]['Total'] += 1

# Crear una lista de diccionarios con los totales de medallas
data = []
for noc, sports in medals_totals.items():
    for sport, medals in sports.items():
        gold_medals = medals['Gold']
        silver_medals = medals['Silver']
        bronze_medals = medals['Bronze']
        total_medals = gold_medals + silver_medals + bronze_medals
        
        data.append({
            'NOC': noc,
            'Sport': sport,
            'Total Gold Medals': gold_medals,
            'Total Silver Medals': silver_medals,
            'Total Bronze Medals': bronze_medals,
            'Total Medals': total_medals
        })

# Crear un nuevo dataframe con los totales de medallas
df_medals_totals = pd.DataFrame(data)

# Imprimir el nuevo dataframe con los totales de medallas
print(df_medals_totals)

df_medals_totals.to_csv('medallas_paises.csv', index=False)
```
